### W poniższym ćwiczeniu zadaniem było stworzenie podanej funkcji, oraz wyrysowanie jej w poszczególnych przedziałach. Całość została wykonana za pomocą biblioteki matplotlib. Stworzylem wyrażenie lambda które zwraca funkcje x^2 + 5, a następnie stworzyłem funkcje która pomaga w rysowaniu wykresów. Przyjmuje ona granice w których ma być wyrysowany wykres, wcześniej utworzony wektor argumentów i zwraca gotowy wykres. Zmieniłem też styl wykresów za pomocą komendy plt.style.use, gdyż w mojej opinii wykresy wtedy lepiej wyglądają.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fun = lambda x: x**2 + 5
```

```python
x = np.linspace(-10,10,1000)
```

```python
plt.style.use('ggplot')
```

```python
def plot_func(a,b,x):
    plt.title(f'Wykres funkcji x^2 + 5 w przedziale [{a},{b}]')
    plt.plot(x,fun(x),'m',label='x^2 + 5')
    plt.xlim(a,b)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    
```

```python
plot_func(-1,1,x)
```

```python
plot_func(-6,6,x)
```

```python
plot_func(0,5,x)
```
